### Рейтинги от Rotten Tomatoes

Начнём с рейтингов Rotten Tomatoes. Здесь два набора рейтингов - рейтинги, опубликованные официальными критиками, и пользовательские рейтинги. 

**Нарисуем график scatterplot, изображающий для Rotten Tomatoes связь между рейтингами от критиков и пользовательскими рейтингами.**

```python

plt.figure(figsize=(10,4),dpi=200)
plt.rcParams.update({'font.size': 8})
plt.ylim(0,100)
plt.xlim(0,100)
sns.scatterplot(x='RottenTomatoes', y='RottenTomatoes_User', data=all_sites)
```

Изучим численную разницу между этими двумя рейтингами. Мы посчитаем эту разницу с помощью формулы RottenTomatoes-RottenTomatoes_User. Полученные значения Rotten_Diff - это рейтинг критиков минус рейтинг пользователей. Близкие к нулю значения говорят о том, что критики и пользователи согласны друг с другом. Большие положительные значения - когда критики оценили фильм намного выше, чем пользователи. А большие отрицательные значения - наоборот, пользователи оценили фильм намного выше, чем критики.

**Создадим новую колонку, в которой сохраним разницу между рейтингом от критиков и пользовательским рейтингом для Rotten Tomatoes. Используем формулу RottenTomatoes-RottenTomatoes_User.**

```python

all_sites['Rotten_Diff']=(all_sites['RottenTomatoes']-all_sites['RottenTomatoes_User']).round(1)
all_sites.head()
```

Теперь посчитаем среднее значение полученной разницы по всему набору данных. Поскольку разница может быть как положительной, так и отрицательной, то начала возьмём абсолютные величины всех значений, и только после этого посчитаем среднее значение. Таким образом мы получим среднее значение абсолютной разницы между рейтингами от критиков и пользовательскими рейтингами.

**Вычислим среднюю абсолютную разницу между рейтингами от критиков и пользовательскими рейтингами для Rotten Tomatoes, согласно описанию выше.**

```python

all_sites['Rotten_Diff'].apply(abs).mean()
```

**Нарисуем график распределения разницы между рейтингами от критиков и пользовательскими рейтингами для Rotten Tomatoes. На этом графике будут отрицательные значения. Для отображения этого распределения используем как KDE, так и гистограммы.**

```python

plt.figure(figsize=(10,4), dpi=200)
sns.histplot(data=all_sites,x='Rotten_Diff',kde=True, bins=25)
plt.rcParams.update({'font.size': 5})
plt.title('RottenTomatoes minus RottenTomatoes_User')
```

**Далее нарисуем график распределения *абсолютных значений* разницы между рейтингами от критиков и пользовательскими рейтингами для Rotten Tomatoes.**

```python

plt.figure(figsize=(10,4), dpi=200)
sns.histplot(data=all_sites,x=abs(all_sites['Rotten_Diff']),kde=True, bins=25)
plt.rcParams.update({'font.size': 20})
plt.title('Abs RottenTomatoes minus RottenTomatoes_User')
```

**выясним, какие фильмы наиболее отличаются по рейтингам. Сначала выведим 5 фильмов с наибольшей по модулю *отрицательной* разницей рейтингов. Поскольку мы вычисляли разницу рейтингов по формуле Critics Rating - Users Rating, большие по модулю отрицательные значения говорят о том, что пользователи в среднем оценили эти фильмы намного выше, чем критики.**

**Какие 5 фильмов в среднем были выше всего оценены пользователями, по сравнению с оценками от критиков:**

```python

all_sites.nsmallest(5, 'Rotten_Diff')[['FILM', 'Rotten_Diff']]
```

**Далее наоборот - отобразим 5 фильмов, которые в среднем были выше всего оценены критиками, по сравнению с оценками от пользователей.**

### Нормализация колонок до диапазона 0-5 в соответствии с колонками STARS и RATINGS от Fandango

Что компании Rotten Tomatoes, Metacritic и IMDB используют другую шкалу измерения рейтингов, чем шкала от 0 до 5 звезд в Fandango. Чтобы выполнить сравнения, нам нужно *нормализовать* эти значения, чтобы они попадали с диапазон от 0 до 5 звёзд.

**Добавим дополнительные колонки в all_sites, и запишите в них нормализованные значения рейтингов от 0 до 5.**

```python
fandango.head()
```

```python
def normalizator(df, fandango_diff_max_min, feature_names):
    result = df.copy()
    for feature_name in feature_names.keys():
        diff_max_min_value = df[feature_name].max() #-df[feature_name].min()
        result[feature_names[feature_name]] = (df[feature_name]* fandango_diff_max_min/ diff_max_min_value).round(1)
    return result
```

```python
feature_names={'RottenTomatoes_User':'RTU_Norm', 'Metacritic_User':'Meta_U_Norm', 'IMDB':'IMDB_Norm'}
fandango_diff_max_min=fandango['STARS'].max() #-fandango['STARS'].min()
all_sites=normalizator(all_sites, fandango_diff_max_min, feature_names)    
```

```python
feature_names={'RottenTomatoes':'RT_Norm', 'Metacritic':'Meta_Norm'}
fandango_diff_max_min=fandango['RATING'].max()#-fandango['RATING'].min()
all_sites=normalizator(all_sites, fandango_diff_max_min, feature_names)    
```

```python
all_sites.head()
```

**Далее создадим DataFrame с названием norm_scores, в котором оставим только нормализованные рейтинги. Также оставим колонки STARS и RATING из исходной таблицы fandango.**

```python

all_sites.columns
```

```python
norm_scores=all_sites[['STARS','RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm']]
```

```python
norm_scores.head()
```
